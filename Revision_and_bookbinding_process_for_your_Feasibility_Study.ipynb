```python
# Graphviz is usually pre-installed in Colab.
# If you run into an error, uncomment the next line to install it:
# !pip install graphviz

import graphviz
from IPython.display import display

# Initialize the flowchart diagram
dot = graphviz.Digraph(comment='Feasibility Study Revision and Bookbind Process')
dot.attr(rankdir='TB', size='8,12') # TB means Top-to-Bottom layout

# Set global default styles for nodes
dot.attr('node', shape='box', style='filled', fillcolor='#E1F5FE', fontname='Arial', fontsize='12')
dot.attr('edge', fontname='Arial', fontsize='10')

# --- Define the Nodes (Steps) ---

# Start and End states (using ovals)
dot.node('Start', 'Start: FS Revision &\nBookbinding Process', shape='oval', fillcolor='#EEEEEE')
dot.node('Finish', 'End: Clearance Complete ✅', shape='oval', fillcolor='#C8E6C9')

# Process steps
dot.node('Step1', '1. Consult Adviser')
dot.node('Step2', '2. Revise Manuscript')
dot.node('Step4', '4. Print 2 Copies on A4 Paper')
dot.node('Step5', '5. Obtain Signatures Before Bookbinding:\n(Adviser ➔ Panelists x2 ➔ Chairman)')
dot.node('Step7', '7. Submit Printed & Signed Copy\nfor Final Checking (FS/E-commerce Prof)')
dot.node('Step8', '8. Upload Soft Copy (PDF + Word)\nto MS Teams')
dot.node('Step9', '9. Bookbind (Follow FS Color Scheme)')
dot.node('Step10', '10. Final Submission to SABM\nBusiness Admin Dept for LRC Endorsement')

# Decision points (using diamonds)
dot.attr('node', shape='diamond', fillcolor='#FFF9C4')
dot.node('Step3', '3. Approved by\nALL Panelists?')
dot.node('Step6', '6. Turnitin Check passed?\n(Plagiarism ≤15%, AI ≤10%)')

# Error/Rejection state
dot.attr('node', shape='box', fillcolor='#FFCDD2')
dot.node('Reject', '⚠️ Manuscript Rejected\n(Exceeded Thresholds)')


# --- Define the Edges (Connections) ---
dot.edge('Start', 'Step1')
dot.edge('Step1', 'Step2')
dot.edge('Step2', 'Step3')

# Panelist Approval Branch
dot.edge('Step3', 'Step4', label=' Yes', color='green', fontcolor='green')
dot.edge('Step3', 'Step2', label=' No (Revise)', color='red', fontcolor='red')

dot.edge('Step4', 'Step5')
dot.edge('Step5', 'Step6')

# Turnitin Branch
dot.edge('Step6', 'Step7', label=' Yes (Passed)', color='green', fontcolor='green')
dot.edge('Step6', 'Reject', label=' No (Failed)', color='red', fontcolor='red')

# Rejection Loop
dot.edge('Reject', 'Step2', label=' Re-revise')

# Final Steps
dot.edge('Step7', 'Step8')
dot.edge('Step8', 'Step9')
dot.edge('Step9', 'Step10')
dot.edge('Step10', 'Finish')

# Render the graph in the Colab output
display(dot)
```
